```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('preprocessed_crime_data.csv')

# Convert 'DATE OCC' to datetime and extract 'Year' and 'Month'
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])
df['Year'] = df['DATE OCC'].dt.year
df['Month'] = df['DATE OCC'].dt.month

# Group by 'Year', 'Month', 'Rpt Dist No' and sum 'Crime Count'
monthly_zone_crime = df.groupby(['Year', 'Month', 'Rpt Dist No'])['Crime Count'].sum().reset_index()

# Define features and target
feature_cols = ['Year', 'Month', 'Rpt Dist No']
X = monthly_zone_crime[feature_cols]
y = monthly_zone_crime['Crime Count']

# Train/test split
train_mask = X['Year'] < 2023
X_train, y_train = X[train_mask], y[train_mask]
X_test, y_test = X[~train_mask], y[~train_mask]

# Parameters from the table
parameters = [
    {"n_estimators": 100, "tree_method": "hist", "max_depth": 3, "learning_rate": 0.01},
    {"n_estimators": 300, "tree_method": "hist", "max_depth": 7, "learning_rate": 0.01},
    {"n_estimators": 500, "tree_method": "hist", "max_depth": 3, "learning_rate": 0.05},
    {"n_estimators": 200, "tree_method": "auto", "max_depth": 5, "learning_rate": 0.05},
    {"n_estimators": 100, "tree_method": "hist", "max_depth": 10, "learning_rate": 0.2},
    {"n_estimators": 400, "tree_method": "hist", "max_depth": 5, "learning_rate": 0.05},
    {"n_estimators": 200, "tree_method": "hist", "max_depth": 3, "learning_rate": 0.05},
    {"n_estimators": 500, "tree_method": "hist", "max_depth": 7, "learning_rate": 0.01},
    {"n_estimators": 300, "tree_method": "auto", "max_depth": 5, "learning_rate": 0.1},
    {"n_estimators": 250, "tree_method": "hist", "max_depth": 6, "learning_rate": 0.07}
]

# Loop through each set of parameters
for idx, params in enumerate(parameters, 1):
    print(f"\nModel {idx} with parameters: {params}")

    # Initialize and train XGBoost model with current parameters
    xgb_regressor = XGBRegressor(
        objective='reg:squarederror',
        random_state=42,
        n_estimators=params['n_estimators'],
        tree_method=params['tree_method'],
        max_depth=params['max_depth'],
        learning_rate=params['learning_rate']
    )
    xgb_regressor.fit(X_train, y_train)

    # Make predictions
    y_pred = xgb_regressor.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    print(f"Mean Squared Error (MSE): {mse:.4f}")

    # Identify the most dangerous zone
    X_test_copy = X_test.copy()
    X_test_copy['Predicted Crime Count'] = y_pred
    zone_preds = X_test_copy.groupby('Rpt Dist No')['Predicted Crime Count'].sum().reset_index()
    most_dangerous = zone_preds.sort_values(by='Predicted Crime Count', ascending=False).iloc[0]
    print("\nMost Dangerous Zone (Predicted):")
    print(f"Rpt Dist No: {most_dangerous['Rpt Dist No']}")
    print(f"Predicted Crime Count: {most_dangerous['Predicted Crime Count']:.4f}")

    # Visualize top 10 most dangerous districts
    top10 = zone_preds.sort_values(by='Predicted Crime Count', ascending=False).head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(data=top10, x='Predicted Crime Count', y=top10['Rpt Dist No'].astype(str), palette='Reds_r')
    plt.title(f'Top 10 Most Dangerous Districts (Model {idx})')
    plt.xlabel('Total Predicted Crime Count')
    plt.ylabel('District Number (Rpt Dist No)')
    plt.tight_layout()
    plt.show()
```
